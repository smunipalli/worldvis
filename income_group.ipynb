```python
import pandas as pd
import wbgapi as wb
import json
```

```python
df = pd.read_csv("./datasets/OGHIST_Mod.csv")
```

```python
def income_groups(x):
    if (x == "L"):
        return "LIC"
    elif (x=="LM"):
        return "LMC"
    elif (x=="UM"):
        return "UMC"
    elif (x=="H"):
        return "HIC"
    else:
        return "INX"
    
```

```python
for k in range(1987, 2022):
    df[str(k)] = df[str(k)].apply(income_groups)
```
