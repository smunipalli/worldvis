```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./datasets/API_SP.POP.2529.MA.5Y_DS2_en_csv_v2_4901832.csv")
```

```python
df
```

```python
for k in range(1960, 2022):
    df[str(k)] = df[str(k)].apply( lambda x: round(x,2) )
```

```python
df
```

```python
df.to_json("percent_male_population_25-29.json")
```
