```python
def getMonth(year, weeknumber):
    date = str(year) + '-W' + str(weeknumber)
    res = datetime.strptime(date + '-1', "%Y-W%W-%w")
    return res.month
    
```

```python
df["Month"] = df.apply(lambda row: getMonth(row.Year, row.Week_Reported) , axis=1)
```
